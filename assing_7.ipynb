```python
fig, ax = plt.subplots(1, 1)

x = np.linspace(stats.norm.ppf(0.01),
                stats.norm.ppf(0.99), 100)
ax.plot(x, stats.norm.pdf(x), 'r-', lw=5, alpha=0.6, label='norm pdf')
r = stats.norm.rvs(size=1000)
ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
```

```python
fig, ax = plt.subplots()
x= np.arange(-4,4,0.001)
ax.plot(x, stats.norm.pdf(x))
ax.set_title("Cumulative normal distribution")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)
px=np.arange(-4,1,0.01)
ax.set_ylim(0,0.5)
ax.fill_between(px,stats.norm.pdf(px),alpha=0.5, color='g')
ax.text(-1,0.1,"cdf(x)", fontsize=20)
plt.show()
```

```python
#finding z score 
def fing_z_score(p, m=0, std=1):
    """
    p=probability
    m= mean
    std= standard devision
    """
    return -stats.norm.ppf(p, m, std)

```

```python
fing_z_score(0.0034)
```
